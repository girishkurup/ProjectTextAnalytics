```python
data=open('C:/Users/Girish.Kurup/OneDrive - Canon EMEA/BIG DATA/1. ITRP Analysis/23jan 2019/TOP incidents 23 Jan2019/test3.csv',encoding='utf-8')
data=data.read().lower()
tokens=word_tokenize(data)
```

```python
dataonlyalpha=[word for word in tokens if word.isalpha()]
```

```python
stop_words = set(stopwords.words('english'))
```

```python
dataonlphanostopwords=[w for w in dataonlyalpha if not w in stop_words]
```

```python
dataonlphanostopwords
```

```python
freq={}
for word in dataonlphanostopwords:
    count=freq.get(word,0)
    freq[word]=count + 1
```

```python
for word in freq.keys():
    print(word + ' : ' + str(freq[word]))
```

```python
results = []
for word in freq.keys():
    tuple = (word, freq[word])
    results.append(tuple)
byFreq=sorted(results, key=lambda word: word[1], reverse=True)

 
```

```python
words_names=[]
words_count=[]
for (word, freq) in byFreq:
    print (word, freq)
    words_names.append(word)
    words_count.append(freq)  
```
